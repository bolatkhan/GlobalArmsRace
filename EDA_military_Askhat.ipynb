```python
# Assuming your dataframe is named 'df'
# First, let's extract the Afghanistan military expenditure data
afg_data = world_bank_data_dropped[world_bank_data_dropped['Country Name'] == 'Afghanistan'].iloc[0]

# Get the years and values
years = [str(year) for year in range(1960, 2024)]  # All years from 1960 to 2023
values = []

for year in years:
    if year in afg_data.index:
        val = afg_data[year]
        # Convert to float if it's a valid number, otherwise None
        if isinstance(val, str) and val.strip() == '..':
            val = None
        else:
            try:
                val = float(val) if val is not None else None
            except (ValueError, TypeError):
                val = None
    else:
        val = None
    values.append(val)
    
# Create a new dataframe for plotting
plot_data = pd.DataFrame({
    'Year': years,
    'Military Expenditure (% of govt)': values
})

# Convert Year to numeric and drop missing values
plot_data['Year'] = pd.to_numeric(plot_data['Year'])
plot_data = plot_data.dropna()

```

```python
type(plot_data)
```

```python
from matplotlib.ticker import MaxNLocator

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=plot_data,
    x='Year',
    y='Military Expenditure (% of govt)',
    marker='o',
    color='darkred'
)

# Force integer ticks (auto-optimized but whole numbers)
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))

plt.title('Afghanistan Military Expenditure (% of general government expenditure)')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Make a copy of the original DataFrame
df = world_bank_data_dropped.copy()

# Step 1: Replace ".." with pd.NA (Pandas' native NA value) and drop all-NA rows
df = df.replace('..', pd.NA).dropna(how='all')

# Step 2: Convert numeric columns to float
# Identify year columns (assuming they're all 4-digit numbers)
year_cols = [col for col in df.columns if isinstance(col, str) and col.isdigit() and len(col) == 4]

# Convert these columns to numeric using pd.to_numeric
for col in year_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 3: Drop irrelevant rows (non-country rows)
# Remove rows where Country Name contains metadata or is NA
df = df[
    df['Country Name'].notna() & 
    ~df['Country Name'].astype(str).str.contains(
        'Data from database|Last Updated', 
        na=False
    )
]

# Reset index after cleaning
df = df.reset_index(drop=True)

# Show the cleaned DataFrame
print(df.head())
print(f"\nCleaned DataFrame shape: {df.shape}")
```

### We got the data in a dataframe. Now it needs to be loaded into our DB.


```python
# we will need the credentials we saved in the .env file
from dotenv import dotenv_values

# We also will need SQLAlchemy and its functions
from sqlalchemy import create_engine, types
from sqlalchemy.dialects.postgresql import JSON as postgres_json

import pandas as pd

# with 'datetime' we want to catch the timestamp of the API call. For the actuality reference. 
# and 'time' for slowing down a .bit
from datetime import datetime
import time
```

```python
# getting API and DB credentials - Alternative 1: dotenv_values()

config = dotenv_values()
 
pg_user = config['POSTGRES_USER'] # align the key labels with your .env file
pg_host = config['POSTGRES_HOST']
pg_port = config['POSTGRES_PORT']
pg_db = config['POSTGRES_DB']
pg_schema = config['POSTGRES_SCHEMA']
pg_pass = config['POSTGRES_PASS']
```

```python
# updating the url
url = f'postgresql://{pg_user}:{pg_pass}@{pg_host}:{pg_port}/{pg_db}'

# creating the engine
engine = create_engine(url, echo=False)
```

```python
# defining data types for the DB
worldbank_dtypes = {
    'Country Name ': types.String,
    'Country Code': types.String,
    'Series Name': types.String,
    **{str(year): types.Float() for year in range(1960, 2024)}
}
```

```python
# writing dataframe to DB
worldbank_cleaned.to_sql(name = 'worldbank_cleaned_new', 
                       con = engine, 
                       schema = pg_schema, # pandas is allowing to specify, in which schema the table shall be created
                       if_exists='replace', 
                       dtype=worldbank_dtypes,
                       index=False
                      )
```
